```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
sns.set_context('paper')
sns.set_style('ticks')
fig, ax = plt.subplots(figsize=(4, 2))
sns.lineplot(df, x='time', y='utilisation', marker='.', markersize=10.)
ax.set_xlabel('Time (mins)')
ax.set_ylabel('A100 Utilisation (%)')
ax.yaxis.set_major_formatter(PercentFormatter(100.))
ax.set_ylim([0., 100.])
ax.set_xlim([0., 20.])
ax.axhspan(80, 90., color='green', alpha=.2)
ax.text(1.2, 110., 'Epoch 1')
ax.text(7.6, 110., 'Epoch 2')
ax.text(13.6, 110., 'Epoch 3')
fig.tight_layout()
```
